```python
import numpy as np
from osgeo import gdal, osr
```

```python
for year in range(2007,2016):
    image=[]
    print('%d'%(year))
    for doy in np.arange(-90,181,1):
        if doy==0:
            continue
        print('.',end='')
        file='D:/hh/@进行中的工作/ESSD_Wheat_Multi-vars/结果/V2-3/tif/%d/%s_Daily_%d%03d.tif'%(year,'TAGP',year,doy) 
        ds = gdal.Open(file)
        data = ds.ReadAsArray()
        gt = ds.GetGeoTransform()
        rows=ds.RasterYSize
        cols=ds.RasterXSize
        image.append(data)
    print('')
    new_file='D:/hh/@进行中的工作/ESSD_Wheat_Multi-vars/结果/V2-3/tif/DEFLATE压缩/CHN_Winter_Wheat_AGB_%d.tif'%(year) 
    driver = gdal.GetDriverByName ( "GTiff" ) # Get a handler to a driver
    dataset_y = driver.Create ( new_file,cols, rows,len(image), gdal.GDT_Int16, options=['COMPRESS=DEFLATE'] )
    dataset_y.SetGeoTransform ( gt)
    dataset_y.SetProjection ( ds.GetProjectionRef() )
    for i in range(len(image)):
        dataset_y.GetRasterBand(i+1).SetNoDataValue(0)
        dataset_y.GetRasterBand(i+1).WriteArray(image[i])
    dataset_y.FlushCache()
    del dataset_y
    print('')

```
